## Background Theory

Harmonograf adalah perangkat mekanik yang menggunakan gerakan pendulum untuk menciptakan pola geometris.

Gerakan tersebut dapat dimodelkan secara matematis menggunakan fungsi sinus teredam:

x(t) = A₁ sin(f₁ t) e^{-d₁ t} + A₂ sin(f₂ t) e^{-d₂ t}  
y(t) = A₃ sin(f₃ t) e^{-d₃ t} + A₄ sin(f₄ t) e^{-d₄ t}

Where:
- A = amplitude
- f = frequency
- d = damping factor
- t = time

**Reference:**  
https://en.wikipedia.org/wiki/Harmonograph

## Interpretation

Persamaan ini menunjukkan bahwa harmonograf pada dasarnya merupakan kombinasi dari beberapa osilasi teredam.

Dengan menyesuaikan frekuensi dan redaman, kita dapat mensimulasikan berbagai perilaku fisik dan menghasilkan pola yang kompleks.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Parameters
Tentukan rentang waktu dan konstanta harmonograf

```python
n = 100000
t = np.logspace(np.log10(10), np.log10(500), n)

A = [2, 2, 1.5, 0.5]     # amplitudo
d = [0.002, 0.001, 0.003, 0.004]  # faktor redaman
f = [1, 1, 5, 0.5]       # frekuensi
```

## Harmonograph Equations

```python
x = A[0]*np.sin(t*f[0])*np.exp(-d[0]*t) + A[1]*np.sin(t*f[1])*np.exp(-d[1]*t)
y = A[2]*np.sin(t*f[2])*np.exp(-d[2]*t) + A[3]*np.sin(t*f[3])*np.exp(-d[3]*t)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x, y, 'k', linewidth=0.2)
plt.axis('off')
plt.title("Harmonograph Pattern")
plt.show()
```

## Osilasi Teredam Tunggal

Grafik ini menunjukkan gelombang sinus teredam sederhana.

Fungsi sinus mewakili osilasi, sementara suku eksponensial mengurangi amplitudo seiring waktu (efek peredaman).

```python
plt.plot(np.sin(t * f[0]) * np.exp(-d[0] * 0.5 * t), linewidth=0.5)
plt.title("Gelombang Sinus Teredam")
plt.xlabel("Waktu")
plt.ylabel("Amplitudo")
plt.show()
```

## Insight

- Peningkatan frekuensi menghasilkan laju osilasi yang lebih tinggi, yang menyebabkan pola yang lebih rumit dan padat karena perubahan arah yang cepat.

- Faktor redaman mengontrol laju kehilangan energi dalam sistem. Redaman yang lebih tinggi menyebabkan gerakan meluruh lebih cepat, menghasilkan pola yang cepat menyusut ke arah pusat.

- Harmonograf pada dasarnya adalah superposisi dari beberapa gerakan osilasi teredam, menjadikannya contoh praktis bagaimana fungsi matematika sederhana dapat memodelkan sistem fisik yang kompleks.

- Perubahan kecil pada parameter (frekuensi dan redaman) dapat secara signifikan mengubah pola akhir, menunjukkan sensitivitas sistem dinamis.

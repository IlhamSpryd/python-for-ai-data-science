## Experiment Log

### Experiment 1: Numerical Data Types (Integers & Floats)
In AI and robotics, integers are often used for counting iterations or discrete items, while floats are crucial for continuous sensor readings (like temperature or distance).

```python
# Initializing discrete and continuous variables
iteration_count = 892003          # Integer
lidar_distance_meters = 892.003   # Float

# Verifying the data types
print("Type of iteration_count:", type(iteration_count))
print("Type of lidar_distance:", type(lidar_distance_meters))
```

> **Observation:** The `type()` function confirms that Python dynamically assigns `int` and `float` classes based on the assigned values. No explicit type declaration is needed.

### Experiment 2: Dynamic String Formatting (f-strings)
Logging system statuses is a daily task in software engineering. Here, we test Python's f-strings to inject numerical variables directly into text.

```python
# Robot state variables
robot_model = "NusantaraBot-X"
battery_level = 89.3

# Formatting the output log using f-string
status_log = f"[SYSTEM INFO] {robot_model} is operational with battery at {battery_level}%."
print(status_log)
```

> **Observation:** Using the `f"..."` syntax allows for clean, readable, and highly efficient string concatenation without needing manual type conversions.

### Experiment 3: String Indexing and Slicing
Text data often contains mixed information. Slicing allows us to extract only the necessary parts. Python uses zero-based indexing.

#### Simulating a raw data string received from a sensor
raw_sensor_data = "ERROR_CODE:404-OBSTACLE_DETECTED"

#### Extracting only the error code (characters from index 11 to 14)
error_code = raw_sensor_data[11:14]

#### Extracting the status message (from index 15 to the end)
status_message = raw_sensor_data[15:]

print("Extracted Code:", error_code)
print("Extracted Message:", status_message)

> **Observation:** String slicing `[start:stop]` is highly precise. The `stop` index is exclusive. This zero-based indexing logic will be critical when moving on to multidimensional arrays (like NumPy arrays or Tensors) in the future.
